```python
import numpy as np

from metrics import (
    calculate_cagr,
    calculate_sharpe_ratio,
    calculate_sortino_ratio,
    calculate_alpha_beta,
    calculate_max_drawdown,
    calculate_tracking_error
)
```

```python
np.random.seed(42)

fund_returns = np.random.normal(
    0.001,
    0.02,
    1000
)

benchmark_returns = np.random.normal(
    0.0008,
    0.018,
    1000
)
```

```python
print(
    "Sharpe:",
    calculate_sharpe_ratio(
        fund_returns
    )
)

print(
    "Sortino:",
    calculate_sortino_ratio(
        fund_returns
    )
)

print(
    "Alpha/Beta:",
    calculate_alpha_beta(
        fund_returns,
        benchmark_returns
    )
)
```

```python
nav = np.cumprod(
    1 + fund_returns
) * 100

print(
    "Max Drawdown:",
    calculate_max_drawdown(nav)
)
```
